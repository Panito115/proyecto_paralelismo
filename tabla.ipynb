```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos en orden (antes, 1, 2, 3, 4)
data = [
    {"Caso": "Antes", "Descarga (s)": 206.04, "Procesamiento (s)": 19.55, "Total (s)": 225.59},
    {"Caso": "1", "Descarga (s)": 32.64, "Procesamiento (s)": 19.40, "Total (s)": 52.04},
    {"Caso": "2", "Descarga (s)": 23.17, "Procesamiento (s)": 4.12, "Total (s)": 27.29},
    {"Caso": "3", "Descarga (s)": 21.69, "Procesamiento (s)": 21.42, "Total (s)": 21.96},
    {"Caso": "4", "Descarga (s)": 21.07, "Procesamiento (s)": 21.34, "Total (s)": 22.10},
]

# Crear DataFrame
df = pd.DataFrame(data)

# Crear figura
fig, ax = plt.subplots(figsize=(7, 2.2))
ax.axis('off')  # Quita ejes

# Crear tabla visual
tabla = ax.table(
    cellText=df.values,
    colLabels=df.columns,
    cellLoc='center',
    loc='center'
)

# Formato de tabla
tabla.auto_set_font_size(False)
tabla.set_fontsize(10)
tabla.scale(1.2, 1.5)  # Ajusta tamaño de tabla

# Título
plt.title("Resumen de Tiempos", fontsize=12, fontweight='bold', pad=10)

# Mostrar tabla
plt.show()
```
